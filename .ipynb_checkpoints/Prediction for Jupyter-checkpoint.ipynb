DHClassifier can accept SMILES formatted molecules as either a single string or a list of strings.

```python
smiles_string='O=C(O)CNCP(=O)(O)O' #glyphosphate
list_of_smiles=['O=C(O)CNCP(=O)(O)O','CC(C)Cc1ccc(cc1)[C@@H](C)C(=O)O'] #glyphosphate and ibuprofen
```

In either case, it will return a dataframe with the predicted probabilities from the four models:  
XG = XGBoost  
LR = Logistic Regression  
RF = Random Forrest  
SVM = Support Vector Machine  
After the class probabilities, it also returns the descriptors as described in the supplementary information.

```python
DHClassifier(smiles_string)
```

```python
DHClassifier(list_of_smiles)
```

You can also request the features to be standardized against the training population using the Standardize=True argument

```python
DHClassifier(smiles_string, Standardize=True)
```
